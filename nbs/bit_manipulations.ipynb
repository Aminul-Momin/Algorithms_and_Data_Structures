#### <font style="color:tomato">Arithmatic vs Logical Right Shift:</font>

In Python, the right shift operation (`>>`) is used for bitwise shifting. However, Python does not differentiate between logical right shift and arithmetic right shift like some other programming languages do. Instead, the behavior of the right shift operation in Python depends on the type of the number being shifted.

- For unsigned integers (int or long in Python 2, and int in Python 3), the right shift operation (`>>`) performs a logical right shift. It shifts the bits to the right and fills the vacant bits on the left with zeros.

````python
bin(8), bin(8 >> 1), int(8 >> 1) # Output: 0b100
````

- For signed integers (int in Python 2 and Python 3), the right shift operation (`>>`) performs an arithmetic right shift. However, Python uses a different mechanism to handle the sign bit. It uses the two's complement representation, so the sign bit is preserved during the right shift operation.

````python
bin(-8), bin(-8 >> 1), int(-8 >> 1)  # Output: -0b1000 (-4 in decimal)
````

### Bit Counts - [EPI: BootCamp]


Write a function `count_bits(x)` which counts the number of bits that are
set to 1 in the given intger, `x`.

- count_bits(0)  == count_bits(int(0b0))        -> 1
- count_bits(1)  == count_bits(int(0b1))        -> 1
- count_bits(2)  == count_bits(int(0b11))       -> 2
- count_bits(15) == count_bits(int(0b1111))     -> 4


````python
def count_bits(x):
    """Counts the number of set-bits in the given integer, 'x'.
    Args:
        x (int): The integer of which set-bits to be counted.
    
    Returns:
        (int): The number of set-bits in 'x'.
    """
    num_bits = 0
    while x:
        num_bits += x & 1
        x >>= 1
    return num_bits
````

### Computing the parity of a word. [EPI:1] 

The parity of a binary word is 1 if the number of 1s in the word is odd; 
otherwise, it is 0. For example, the parity of 101 is 1, and the parity of 
10001000 is 0.

Write a function 'parity(x)' that computes the parity of a single 64-bit 
word.

- Examples:
    1. parity() -> 

````python
def parity(x):
    """Computes the parity of a the given word.
    
    Args:
        x (int): 
    
    Returns:
        (int): The parity of binary word.
    """
    result = 0
    while x:
        result ^= x & 1
        x >>= 1
    return result
````

### Swap bits. [EPI:2] 

Write a function 'swap_bits(x, i, j)' which takes as input a 64-bit integer
and swaps the bits at indices 'i' and 'j'.

- Examples:
    - swap_bits(x, i, j)

````python
def swap_bits(x, i, j):
    # Extract the i-th and j-th bits, and see if they differ.
    if (x >> i) & 1 != (x >> j) & 1:
        # i-th and j-th bits differ. We will swap them by flipping their values.
        # Select the bits to flip with bit_mask. Since x^1 = 0 when x = 1 and 1
        # when x = 0, we can perform the flip XOR.
        bit_mask = (1 << i) | (1 << j)
        x ^= bit_mask
    return x
````

### Find a closest integer with the same weight. [EPI:4] 

Define the weight of a nonnegative integer x to be the number of bits that 
are set to 1 in its binary representation. For example, since 92 in 
base-2 equals (1011100), the weight of 92 is 4.

Write a program 'closest_int_same_bit_count(x)' which takes as input a 
nonnegative integer x and returns a number y which is not equal to x, but 
has the same weight as x and their difference |y-x|, is as small as 
possible. You can assume x is not 0, or all 1s.

- Examples:
    1. closest_int_same_bit_count(6) -> 5

````python
def closest_int_same_bit_count(x):

    NUM_UNSIGNED_BITS = 64
    for i in range(NUM_UNSIGNED_BITS - 1):
        if (x >> i) & 1 != (x >> (i + 1)) & 1:
            x ^= (1 << i) | (1 << (i + 1))  # Swaps bit-i and bit-(i + 1).
            return x

    # Raise error if all bits of x are 0 or 1.
    raise ValueError("All bits are 0 or 1")
````

### Compute x * y without arithmetical operators. [EPI:5] 

```
Write a function 'multiply(x,y)' which takes two integer as inputs and 
computes their product without arithmetical operators.

Examples:
    1. multiply(0,1)   -> 0
    2. multiply(1,1)   -> 1
    3. multiply(2,1)   -> 2
    4. multiply(2,2)   -> 4
    5. multiply(10,10) -> 100
```

````python
def multiply(x, y):
    def add(a, b):
        while b:
            carry = a & b
            a, b = a ^ b, carry << 1
        return a

    running_sum = 0
    while x:  # Examines each bit of x.
        if x & 1:
            running_sum = add(running_sum, y)
        x, y = x >> 1, y << 1
    return running_sum
````

### Compute x/y. [EPI:6] 

```
Write a function 'divide(x,y)' which takes two integer as inputs and 
computes their quotient, using only the addition, subtraction, and 
shifting operator.

Examples:
    1. divide(x, y) -> 
```

````python
def divide(x, y):

    result, power = 0, 32
    y_power = y << power
    while x >= y:
        while y_power > x:
            y_power >>= 1
            power -= 1

        result += 1 << power
        x -= y_power
    return result
````

### Compute $x^y$. [EPI:7] 

```
Write a function 'power(x, y)' which takes a double x and an integer y as inputes and return 'x to the power of y'. You can ignore overflow and underflow.

Examples:
    1. power(x, y) -> 
```

````python
def power(x, y):

    result, power = 1.0, y
    if y < 0:
        power, x = -power, 1.0 / x
    while power:
        if power & 1:
            result *= x
        x, power = x * x, power >> 1
    return result
````

### Reverse digits. [EPI:8] 

```
Write a program which takes an integer and returns the integer 
corresponding to the digits of input written in reverse order.

Examples:
    1. reverse(42) -> 24
    2. reverse(-314) -> -413
```

````python
def reverse(x):

    result, x_remaining = 0, abs(x)
    while x_remaining:
        result = result * 10 + x_remaining % 10
        x_remaining //= 10
    return -result if x < 0 else result
````

### Check if a decimal integer is a palindrome. [EPI:9] 

```
    Write a function 'is_palindrome_number(x)' which takes an integer and 
    determines if that integer's representation as a decimal string is a 
    palindrom.

    Examples:
        1. is_palindrome_number(0)          -> True
        2. is_palindrome_number(11)         -> True
        3. is_palindrome_number(121)        -> True
        4. is_palindrome_number(333)        -> True
        5. is_palindrome_number(2147447412) -> True
        6. is_palindrome_number(-1)         -> False
        7. is_palindrome_number(12)         -> False
```

````python
def is_palindrome_number(x):

    if x <= 0:
        return x == 0

    num_digits = math.floor(math.log10(x)) + 1
    msd_mask = 10 ** (num_digits - 1)
    for i in range(num_digits // 2):
        if x // msd_mask != x % 10:
            return False
        x %= msd_mask  # Remove the most significant digit of x.
        x //= 10  # Remove the least significant digit of x.
        msd_mask //= 100
    return True
````

### Generate uniform random numbers. [EPI:10] 

How would you implement a random number generator that generates a random integer `i` between `a` nad `b`, inclusive, given a random number that produces zero or one with equal probability: All values in `[a,b]` should be equally likely.

````python
def nonuniform_random_number_generation(values, probabilities):

    prefix_sum_of_probabilities = list(itertools.accumulate(probabilities))
    interval_idx = bisect.bisect(prefix_sum_of_probabilities, random())
    return values[interval_idx]
````

### Rectangle intersection. [EPI:11] 

Write a program which tests if two rectangles have a nonempty intersection. If the intersection is nonempty, return the rectangle formed by their intersection.

````python
Rectangle = collections.namedtuple("Rectangle", ("x", "y", "width", "height"))


def intersect_rectangle(R1, R2):
    def is_intersect(R1, R2):
        return (
            R1.x <= R2.x + R2.width
            and R1.x + R1.width >= R2.x
            and R1.y <= R2.y + R2.height
            and R1.y + R1.height >= R2.y
        )

    if not is_intersect(R1, R2):
        return Rectangle(0, 0, -1, -1)  # No intersection.
    return Rectangle(
        max(R1.x, R2.x),
        max(R1.y, R2.y),
        min(R1.x + R1.width, R2.x + R2.width) - max(R1.x, R2.x),
        min(R1.y + R1.height, R2.y + R2.height) - max(R1.y, R2.y),
    )
````

#### Find the missing IP address. [EPI: 11.9]

````python
def find_missing_element(stream):

    NUM_BUCKET = 1 << 16
    counter = [0] * NUM_BUCKET
    stream, stream_copy = itertools.tee(stream)
    for x in stream:
        upper_part_x = x >> 16
        counter[upper_part_x] += 1

    # Look for a bucket that contains less than (1 << 16) elements.
    BUCKET_CAPACITY = 1 << 16
    candidate_bucket = next(
        i for i, c in enumerate(counter) if c < BUCKET_CAPACITY)

    # Finds all IP addresses in the stream whose first 16 bits are equal to
    # candidate_bucket.
    candidates = [0] * BUCKET_CAPACITY
    stream = stream_copy
    for x in stream_copy:
        upper_part_x = x >> 16
        if candidate_bucket == upper_part_x:
            # Records the presence of 16 LSB of x.
            lower_part_x = ((1 << 16) - 1) & x
            candidates[lower_part_x] = 1

    # At least one of the LSB combinations is absent, find it.
    for i, v in enumerate(candidates):
        if v == 0:
            return (candidate_bucket << 16) | i

    raise ValueError('no missing element')
````
